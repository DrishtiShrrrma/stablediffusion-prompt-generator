```python
import pandas as pd
```

```python
prompts = pd.read_csv("prompts.csv")
prompts.head()
```

```python
prompts.prompt.str.split().str.len()
```

```python
clean_prompt = prompts[prompts.prompt.str.split().str.len()>6].reset_index(drop=True)
clean_prompt
```

```python
clean_prompt.to_csv("cleaned_prompts.csv", columns=["prompt"], header=False, index=False )
```

```python
from aitextgen import aitextgen
from aitextgen.TokenDataset import TokenDataset
```

```python
data = TokenDataset('/content/cleaned_prompts.csv', line_by_line=True)
```

```python
model="EleutherAI/gpt-neo-125M"
```

```python
ai = aitextgen(model = model,  to_gpu=True)
```

```python
from pytorch_lightning import loggers
tb_logger = loggers.TensorBoardLogger('logs/')
```

```python
ai.train(data,
         loggers=tb_logger,
         from_cache=False,
         num_steps=500,
         generate_every=100,
         save_every=500,
         save_gdrive=False,
         learning_rate=1e-3,
         fp16=False,
         batch_size=1,
         )
```

```python
ai.save("StableDiffusion-Prompt-Generator-GPT-Neo-125M-k1")
```
